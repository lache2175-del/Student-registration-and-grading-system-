```python
import sys
sys.path.append('C:\\Users\\m5fet\\Desktop\\Project')
import pack1.mod1 as call

L1 = {}

name = input("Enter Your Name: ")
student_id = input("Enter Your College ID: ")
spec = input("What is Your Major?: ")
sem = int(input("Inter your Current Semester: "))

try:
    abs_per = float(input("Enter your Absent percentage: "))
except ValueError:
    print("Warning! Input should be integer")
    abs_per = 0.0

L1.update({"Name": name, "ID": student_id, "Major": spec, "Semester": sem})

call.f1()

print(f"Student: {L1['Name']}")
print(f"ID: {L1['ID']}")
print(f"Semester: {L1['Semester']}")
print(f"Specialization: {L1['Major']}")

call.f2(abs_per)

final_gpa = call.f4()

L1["CGPA"] = final_gpa

for u, o in L1.items():
    print(f"{u} : {o}")

if abs_per > 20 or final_gpa < 2.5:
    print("Result: You didn't pass in this semester")
else:
    print("Result: You passed in this semester")
```
